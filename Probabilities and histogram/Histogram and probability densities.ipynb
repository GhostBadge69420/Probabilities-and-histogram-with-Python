```python
x = np.linspace(-3,3,123)

fig,ax = plt.subplots(2,3)

for i in range(3):

    # create the function and its histogram
    y = x**(i+1)
    h = np.histogram(y)
    yP = h[0]/np.sum(h[0])

    # plotting 
    ax[0,i].plot(x,y)
    ax[0,i].set_title('$y=x^{%g}$' %(i+1))

plt.show()
```

```python
x = np.linspace(-3,3,123)

fig,ax = plt.subplots(2,3)
fig.set_size_inches(10,6)

for i in range(3):

    # create the function and its histogram
    y = x**(i+1)
    h = np.histogram(y)
    yP = h[0]/np.sum(h[0])

    # plotting 
    ax[0,i].plot(x,y)
    ax[0,i].set_title('$y=x^{%g}$' %(i+1))

    ax[1,i].plot(h[1][:-1],yP)

plt.show()
```
