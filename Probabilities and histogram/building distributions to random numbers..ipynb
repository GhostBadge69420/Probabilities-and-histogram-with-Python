```python
n = 500

orig_data = 2**np.random.randn(n)

# rank data
rank_data = rankdata(orig_data)

# normalize to [-1 1]
norm_data = 2*(rank_data/np.max(rank_data))-1

fig,ax = plt.subplots(1,2)

ax[0].hist(orig_data,30)
ax[0].set_xlabel('Value')
ax[0].set_ylabel('Count')

ax[1].hist(norm_data,30)
ax[1].set_xlabel('Value')
ax[1].set_ylabel('Count')

plt.show()
```

```python
x = np.random.randn(1000)
y = 2**x
 
y_rank = sp.stats.rankdata(y)
 
scaler = MinMaxScaler(feature_range=(-0.99,0.99))
y_scaled = scaler.fit_transform(y_rank.reshape(-1,1))
 
y_normalized = np.arctanh(y_scaled)
 
plt.subplot(1,2,1)
sns.histplot(y)
plt.subplot(1,2,2)
sns.histplot(y_normalized);
plt.tight_layout()
```
