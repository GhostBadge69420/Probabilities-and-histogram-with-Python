```python
n = 1000

x = np.random.randn(n)
y = np.random.randn(n)**2

plt.hist2d(x,y,30,vmax=10)
plt.xlabel('x')
plt.ylabel('$x^2$')
plt.show()
```

```python
sns.set()

sns.jointplot(x = np.random.randn(n),y = np.random.randn(n)**2,kind='scatter',color=[.8,.3,.9]).plot_joint(sns.kdeplot)
plt.show()
```

```python
sns.set()

sns.jointplot(x = np.random.randn(n),y = np.random.randn(n)**2\
              ,kind='scatter',color=[.8,.3,.9]).plot_joint(sns.kdeplot)

plt.xlabel('x')
plt.ylabel('$x^2$')

plt.show()
```
